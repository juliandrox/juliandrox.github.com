```python
#Programa Teorema de Pitágoras
cateto_a = int(float(input("Ingrese el valor del cateto adyacente: \t")))
cateto_o = int(float(input("Ingrese el valor del cateto opuesto: \t")))
hipotenusa = ((cateto_a**2) + (cateto_o**2)) ** 0.5
print("El valor de la Hipotenusa es: ", hipotenusa)
```

```python
#Programa Calculo de Potencias

p1 = float(input("Ingrese el valor de la primera potencia: \t"))
p2 = float(input("Ingrese el valor de la segunda potencia: \t"))
resultado = p1**p2
print(resultado)
```

```python
#Programa Radio y Circunferencia de un Círculo
radio = float(input("Ingresa el radio: "))
pi = 3.14
area = pi * radio * radio
circunferencia = 2.0 * pi * radio

print('Radio =', radio)
print('Area =', area)         
print('Circunferencia = ', circunferencia)
```

```python
#Calculo factorial de 5! y 10!

n1 = 5
n2 = 10

Calculo_Factorial = n1 * (n1-1) * (n1-2) * (n1-3) * (n1-4)
print(Calculo_Factorial)

Calculo_Factorial2 = n2 * (n2-1) * (n2-2) * (n2-3) * (n2-4) * (n2-5) * (n2-6) * (n2-7) * (n2-8) * (n2-9)
print(Calculo_Factorial2)
```

```python
#Calculadora números pares e impares

data = int(input("Ingresa un número entero: "))
if data%2 == 0:
    print("El número ", data, " es par")
else: 
    print("El número ", data, " es impar")
```
